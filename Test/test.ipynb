```python
import pandas as pd
from catboost import CatBoostRegressor
import matplotlib.pyplot as plt
```

```python
model = CatBoostRegressor()
model.load_model("araba_fiyat_model.cbm")
```

```python
cd=pd.read_csv("cars.csv")
```

```python
data=cd.head(1000)
```

```python
td = pd.DataFrame(columns=["Tahmin", "Gercek_fiyat"])
```

```python
for i in range(1000):
    test = data[["konum","marka","seri","model","yil","kilometre","vites_tipi","yakit_tipi","kasa_tipi","renk","motor_hacmi","motor_gucu","cekis","ortalama_yakit_tuketimi","yakit_deposu","tramer","degisen","boya"]].iloc[[i]] 
    sonuc = model.predict(test)[0]  
    td.loc[i] = [int(sonuc), cd["fiyat"].iloc[i]]
```

```python
plt.figure(figsize=(10,5))
plt.plot(td["Gercek_fiyat"].head(100), label="Ger√ßek Fiyat")
plt.plot(td["Tahmin"].head(100), label="Tahmin")
plt.legend()
plt.show()
```
