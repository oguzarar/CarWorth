```python
import numpy as np
import pandas as pd
from catboost import CatBoostRegressor
```

```python
cd=pd.read_csv("cars.csv")
```

```python
cd.to_csv("cars.csv")
```

```python
cd.head(50)
```

```python
cd['fiyat'] = cd['fiyat'].str.replace(' TL', '', regex=False).str.replace('.', '', regex=False).astype(float)

cd['kilometre'] = cd['kilometre'].str.replace(' km', '', regex=False).str.replace('.', '', regex=False).astype(float)

cd['ortalama_yakit_tuketimi'] = cd['ortalama_yakit_tuketimi'].str.replace(' lt', '', regex=False).str.replace(',', '.', regex=False).astype(float)

cd['yakit_deposu'] = cd['yakit_deposu'].str.replace(' lt', '', regex=False).astype(float)
```

```python
cd['motor_gucu'] = cd['motor_gucu'].str.replace(' HP', '', regex=False)
cd['motor_gucu'] = cd['motor_gucu'].str.replace(' hp', '', regex=False)
```

```python
cd['motor_hacmi'] = cd['motor_hacmi'].str.replace(' cm3', '', regex=False)
cd['motor_hacmi'] = cd['motor_hacmi'].str.replace(' cc', '', regex=False)
```

```python
bagimsiz_degisken=cd[["konum","marka","seri","model","yil","kilometre","vites_tipi","yakit_tipi","kasa_tipi","renk","motor_hacmi","motor_gucu","cekis","ortalama_yakit_tuketimi","yakit_deposu","tramer","degisen","boya"]]
bagimli_degisken=cd["fiyat"]

kategori=["konum","marka","seri","model","vites_tipi","yakit_tipi","kasa_tipi","renk","motor_hacmi","motor_gucu","cekis"]

model = CatBoostRegressor(cat_features=kategori, verbose=0)
model.fit(bagimsiz_degisken, bagimli_degisken)

test = pd.DataFrame([
    {
        "konum":"Ankara",
        "marka":"BMW",
        "seri":"5 Serisi",
        "model":"520i Premium",
        "yil":2015,
        "kilometre":282000,
        "vites_tipi":"Otomatik",
        "yakit_tipi":"Benzinli",
        "kasa_tipi":"Sedan",
        "renk":"Gri",
        "motor_hacmi":1592,
        "motor_gucu":170,
        "cekis":"Arkadan İtiş",
        "ortalama_yakit_tuketimi":5.5,
        "yakit_deposu":70,
        "tramer":0,
        "degisen":0,
        "boya":0
    }
])
sonuc = model.predict(test)

sonuc
```
